```python
spy = pd.read_csv('SPY2000_2021.CSV', parse_dates = [0] )
spy['Adj Close'].pct_change(5283)
```

```python
spy['Gains'] = spy['Adj Close'].pct_change(1)
```

```python
spy = spy.set_index('Date')
```

```python
spy['Gains'].sort_values(ascending = False).head(10)
```

```python
increase = []
decrease = []
def change(gains):
    if gains>0:
        increase.append(gains)
        return increase
    else:
        decrease.append(gains)
        return decrease
spy['Gains'].apply(change)
increase
```

```python
max_increase = sorted(increase)
```

```python
max_decrease = sorted(decrease)
```
