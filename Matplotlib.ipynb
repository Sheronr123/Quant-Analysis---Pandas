```python
a = np.linspace(0,11,10)
b = np.linspace(0,15,10)
c = np.linspace(0,20,10)
d = np.linspace(0,25,10)
fig, axes = plt.subplots(2,2)
axes[0][0].plot(a,a, label = 'linear', color = 'purple', lw =1, ls='--', marker = 'x', ms = 7)
axes[0][0].plot(a,a**2, label = 'exponential', color = '#cc3110',lw =1)
axes[0][0].legend(loc = 0)

axes[1][0].plot(b,b, label = 'linear')
axes[1][0].plot(b,b**2, label = 'exponential')
axes[1][0].legend(loc = 0)

axes[0][1].plot(c,c, label = 'linear')
axes[0][1].plot(c,c**2, label = 'exponential')
axes[0][1].legend(loc = 0)

axes[1][1].plot(d,d, label = 'linear')
axes[1][1].plot(d,d**2, label = 'exponential')
axes[1][1].legend(loc = 0)

plt.suptitle('Lin vs Expo')
fig.subplots_adjust(wspace = 0.5, hspace = 0.5)
plt.tight_layout()

plt.savefig('subplots1.jpeg', bbox_inches = 'tight')
```

```python
m = np.linspace(0,10,11)
c = 299337984
e = m *(c**2)
import matplotlib.pyplot as plt
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(m,e, label = 'E = MC^2', color = 'red', lw = 3, ls= '--')
axes.set_xlabel('Mass')
axes.set_ylabel('Energy')
axes.legend(loc = 4)
```

```python
labels = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']

july16_2007 =[4.75,4.98,5.08,5.01,4.89,4.89,4.95,4.99,5.05,5.21,5.14]
july16_2020 = [0.12,0.11,0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2,1)
axes[0].plot(labels,july16_2007, label = 'July 2007', lw = 2, marker = 'o', color = 'red')
axes[1].plot(labels,july16_2020,label = 'July 2020', lw = 2, marker = 'o')
axes2 = axes.twinx()
axes2.plot(labels,july16_2020,label = 'July 2020', lw = 2, marker = 'o')
axes[0].legend()
axes[1].legend()
```
