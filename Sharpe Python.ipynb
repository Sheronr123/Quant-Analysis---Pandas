```python
def compute_sharpe(daily_return, risk_free_rate = 0):
     return ((daily_return['Daily Return'].mean()-risk_free_rate)/(daily_return['Daily Return'].std())) 
```

```python
import scipy.stats
```

```python
def compute_pdr(daily_return, bmark = 0):
    sr = compute_sharpe(daily_return)
    skew = scipy.stats.skew(daily_return['Daily Return'])
    kurtosis = scipy.stats.kurtosis(daily_return['Daily Return'], fisher = True)
    n = len(daily_return)
    sigma_sr =  ( (1 / (n-1)) * (1 + 0.5 * sr**2 - skew * sr + (kurtosis / 4) * sr**2))**0.5
    ratio = (sr - bmark)/sigma_sr
    psr = scipy.stats.norm.cdf(ratio)
    return psr
```

```python
compute_pdr(apple)
```

```python
def compute_prob_sharpe_ratio(data, benchmark=0):
    
    sr = compute_sharpe(data, 0)
    
    skew = scipy.stats.skew(data["Daily Return"])
    # Use fisher kurtosis
    
    kurtosis = scipy.stats.kurtosis(data["Daily Return"], fisher=True)  
    
    n = len(data)
    
    std = ( (1 / (n-1)) * (1 + 0.5 * sr**2 - skew * sr + (kurtosis / 4) * sr**2))**0.5
    
    ratio = (sr - benchmark) / std
    prob_sharpe_ratio = scipy.stats.norm.cdf(ratio)
    return prob_sharpe_ratio
```

```python
compute_prob_sharpe_ratio(apple)
```
